```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("covid_data.csv")
data = data[["Province_State","Country_Region","Last_Update","Lat","Long_","Confirmed","Recovered","Deaths","Active"]]
data.columns = ["State","Country","Update","Latitude","Longitude","Confirmed","Recovered","Deaths","Active"]
print(data.head())
```

```python
print(data.info())
```

```python
data['State'] = data['State'].fillna(data['State'].mode()[0])
data['Update'] = data['Update'].fillna(data['Update'].mean())
data['Latitude'] = data['Latitude'].fillna(data['Latitude'].mean())
data['Longitude'] = data['Longitude'].fillna(data['Longitude'].mean())
```

```python
print(data.info())
```

```python
top_ten_countries = pd.DataFrame(data.groupby("Country")["Confirmed"].sum().sort_values(ascending=False).head(10))
top_ten_countries.reset_index(inplace=True)
print(top_ten_countries)
one = px.scatter(top_ten_countries, x="Country", y="Confirmed", size="Confirmed", color="Country",size_max=120,title="Top 10 Countries with Highest Confirmed Cases")
one.show()
one.write_html("top_ten_countries.html")
```

```python
confirmed_cases_by_country = pd.DataFrame(data.groupby("Country")["Confirmed"].sum().sort_values(ascending=False).head(10))
confirmed_cases_by_country.reset_index(inplace=True)
print(confirmed_cases_by_country)
two = px.bar(confirmed_cases_by_country, x="Country", y="Confirmed", color="Country", title="Confirmed Cases by Country")
two.show()
two.write_html("confirmed_cases_by_country.html")
```

```python
top_ten_countries_deaths = pd.DataFrame(data.groupby("Country")["Deaths"].sum().sort_values(ascending=False).head(10))
top_ten_countries_deaths.reset_index(inplace=True)
print(top_ten_countries_deaths)
three = px.bar(top_ten_countries_deaths, x="Country", y="Deaths", color="Country", title="Top 10 Countries with Highest Deaths")
three.show()
three.write_html("top_ten_countries_deaths.html")
```

```python
top_ten_countries_recovered = pd.DataFrame(data.groupby("Country")["Recovered"].sum().sort_values(ascending=False).head(10))
top_ten_countries_recovered.reset_index(inplace=True)
print(top_ten_countries_recovered)
three = px.bar(top_ten_countries_recovered, x="Country", y="Recovered", color="Country", title="Top 10 Countries with Highest Recovered Cases")
three.show()
three.write_html("top_ten_countries_recovered.html")
```

```python
top_states_uk = pd.DataFrame(data[data["Country"]=="United Kingdom"].groupby("State")["Confirmed"].sum().sort_values(ascending=False).head(10))
top_states_uk.reset_index(inplace=True)
print(top_states_uk)
four = px.bar(top_states_uk, x="State", y="Confirmed", color="State", title="Top 10 States in United Kingdom with Highest Confirmed Cases")
four.show()
four.write_html("top_states_uk.html")
```

```python
top_states_brazil = pd.DataFrame(data[data["Country"]=="Brazil"].groupby("State")["Confirmed"].sum().sort_values(ascending=False).head(10))
top_states_brazil.reset_index(inplace=True)
print(top_states_brazil)
five = px.bar(top_states_brazil, x="State", y="Confirmed", color="State", title="Top 10 States in Brazil with Highest Confirmed Cases")
five.show()
five.write_html("top_states_brazil.html")
```

```python
top_states_brazil = data['Country']=='Brazil'
top_five_states_brazil = data[top_states_brazil].nlargest(5, 'Confirmed')
fig= go.Figure(data=[
    go.Bar(name='Confirmed', x=top_five_states_brazil['State'], y=top_five_states_brazil['Confirmed']),
    go.Bar(name='Recovered', x=top_five_states_brazil['State'], y=top_five_states_brazil['Recovered']),
    go.Bar(name='Deaths', x=top_five_states_brazil['State'], y=top_five_states_brazil['Deaths'])
])
fig.update_layout(title='Top 5 States in Brazil with Highest Confirmed Cases', xaxis_title='State', yaxis_title='Count', barmode='group')
fig.show()
fig.write_html("top_five_states_brazil.html")
```
